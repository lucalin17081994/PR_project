```python
data_dir = "data"

directory = os.fsencode(data_dir)
    
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename=='train_split.txt':
        df_train = pd.read_csv(data_dir+"/"+filename, sep=" ", header=None)
        df_train.columns = ["patientid","Image", "label", "dataset"]
    if filename=='test_split.txt':
        df_test = pd.read_csv(data_dir+"/"+filename, sep=" ", header=None)
        df_test.columns = ["patientid","Image", "label", "dataset"]
df = pd.concat([df_train,df_test])
```
